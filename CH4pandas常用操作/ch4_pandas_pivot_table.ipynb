1.去除空值

```python
# 去除空值
test_df = pd.DataFrame({"超商":["7-11","全家","萊爾富"], "上月銷售額(萬)":[30,25,np.NAN],"客戶滿意度":[np.NAN,30,30],"失竊次數":[np.NAN,np.NAN,np.NAN]})
# test_df.dropna(subset=['客戶滿意度'], inplace=True, how='any', axis=0)
# test_df.dropna(subset=['客戶滿意度'], inplace=True, how='all', axis=0)
test_df.dropna(inplace=True, how='all', axis=1)
test_df
```

2.去除重複值

```python
test_df = pd.DataFrame({"超商":["7-11","全家","全家","萊爾富","7-11"], "上月銷售額(萬)":[30,25,25,np.NAN,100],"客戶滿意度":[np.NAN,30,30,30,40],"失竊次數":[np.NAN,np.NAN,np.NAN,np.NAN,np.NAN]})
# 去除重複值
# test_df.drop_duplicates(inplace=True)
test_df.drop_duplicates(subset=["超商"],inplace=True, keep=False)
test_df
```

3.空值填充

```python
test_df = pd.DataFrame({"超商":["7-11","全家","全家","萊爾富","7-11"], "上月銷售額(萬)":[30,25,25,np.NAN,100],"客戶滿意度":[np.NAN,30,30,30,40],"失竊次數":[np.NAN,np.NAN,np.NAN,np.NAN,np.NAN]})
# test_df.fillna(value=0,inplace=True)

test_df.fillna({'失竊次數':0,'客戶滿意度':'讚'},inplace=True)
test_df
```

4.Columns rename

```python
test_df = pd.DataFrame({"超商":["7-11","全家","全家","萊爾富","7-11"], "上月銷售額(萬)":[30,25,25,np.NAN,100],"客戶滿意度":[np.NAN,30,30,30,40],"失竊次數":[np.NAN,np.NAN,np.NAN,np.NAN,np.NAN]})
test_df.rename(columns={"超商":"便利商店"}, inplace=True) # 注意用法
test_df
```

5.排序

```python
 # 資料排序
test_df = pd.DataFrame({"超商":["7-11","全家","全家","萊爾富","7-11"], "上月銷售額(萬)":[30,25,60,100,100],"客戶滿意度":[np.NAN,30,40,60,40],"失竊次數":[np.NAN,np.NAN,np.NAN,np.NAN,np.NAN]})
test_df.sort_values(by=["上月銷售額(萬)","客戶滿意度"],ascending=[True, False], inplace=True, ignore_index=True)
test_df
```

6.資料篩選(mask)

```python
# mask:彈性條件的篩選數據
test_df = pd.DataFrame({"超商":["7-11","全家","全家","萊爾富","7-11"], "上月銷售額(萬)":[30,25,60,100,100],"客戶滿意度":[np.NAN,30,40,60,40],"失竊次數":[np.NAN,np.NAN,np.NAN,np.NAN,np.NAN]})

mask1 = test_df["上月銷售額(萬)"] >=100
mask2 = test_df.超商.str.contains('7')
# test_df
test_df = test_df[~mask1 & ~mask2]
test_df
```

```python
# 樞紐分析，想知道ID與工時的關係
pvt = df.pivot_table(values='WORK_TIME',
                     index=['ID_NO'],aggfunc=np.mean)
pvt.head(10)
```

```python
# 指定統計值(預設是mean)
pvt = df.pivot_table(values='WORK_TIME',
                     index='ID_NO',aggfunc=np.min)

pvt.head(10)
```

```python
# 使用參數columns指定另一維度的分析
pvt = df.pivot_table(values='WORK_TIME',
                     index='ID_NO',
                     columns=['CURRENT_SHOP_CODE'],fill_value=0)
pvt.head()
```
